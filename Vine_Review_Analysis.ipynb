```python
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/vine_table.csv"
vine_table_df = pd.read_csv(csv_path)

vine_table_df.head()
```

```python
vine_table_df.columns = ["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"]

vine_table_df.head()
```

```python
vine_table_df.info()
```

```python
vine_table_votes_df = vine_table_df.loc[vine_table_df['total_votes']>=20]
vine_table_votes_df
```

```python
new_vine_table_df = vine_table_votes_df.loc[((vine_table_votes_df['helpful_votes'])/vine_table_votes_df['total_votes'])>0.5]
new_vine_table_df
```

```python
only_vine_table_df = new_vine_table_df.loc[(new_vine_table_df['vine']=="Y")]
only_vine_table_df
```

```python
only_vine_table_df.count()
```

```python
total_vine_reviews=only_vine_table_df.count()
total_vine_reviews
```

```python
non_vine_table_df = new_vine_table_df.loc[(new_vine_table_df['vine']=="N")]
non_vine_table_df
```

```python
total_non_vine_reviews=non_vine_table_df.count()
total_non_vine_reviews
```

```python
only_vine_table_df.loc[only_vine_table_df['star_rating']==5].count()
```

```python
only_vine_5_reviews = only_vine_table_df.loc[only_vine_table_df['star_rating']==5].count()
only_vine_5_reviews
```

```python
non_vine_table_df.loc[non_vine_table_df['star_rating']==5].count()
```

```python
non_vine_5_reviews = non_vine_table_df.loc[non_vine_table_df['star_rating']==5].count()
non_vine_5_reviews
```

```python
percentage_paid_5s = (only_vine_5_reviews['review_id']/ total_vine_reviews['review_id'])*100
percentage_paid_5s
```

```python
percentage_unpaid_5s = (non_vine_5_reviews['review_id']/ total_non_vine_reviews['review_id'])*100
percentage_unpaid_5s
```
